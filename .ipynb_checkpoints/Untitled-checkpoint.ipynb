```python
import matplotlib.pyplot as plt
import numpy as np

COLOR_MAP = {
    'water': (0.255, 0.298, 0.529),
    'land': (135/255, 86/255, 72/255),
    'grid_field': (0.612, 0.702, 0.831),
    'tidal_flat': (246/255, 225/255, 198/255)
}

# 中文到英文的映射
CHINESE_TO_ENGLISH = {
    '水域': 'water',
    '陆地': 'land',
    '养殖区域': 'grid_field',
    '潮滩': 'tidal_flat'
}

def prepare_plot(data_labels, categories_l):
    # Create an empty RGB image filled with white color
    img_shape = next(iter(data_labels.values())).shape
    img = np.ones((*img_shape, 3))

    # Convert categories_l from Chinese to English
    categories = [CHINESE_TO_ENGLISH[cat] for cat in categories_l]

    # Apply colors to the image based on the categories
    for category in categories:
        mask = data_labels[category]
        color = COLOR_MAP[category]
        img[mask] = color

    # Plot the image
    plt.imshow(img)
    
    # Create legend with squares
    handles = [plt.Line2D([0], [0], marker='s', color='w', markerfacecolor=color, markersize=10, linestyle='None', label=CHINESE_TO_ENGLISH.get(category, category)) 
               for category, color in COLOR_MAP.items() if category in categories]
    
    plt.legend(handles=handles)
    plt.axis('off')
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
from matplotlib.patches import Patch

# 超参数设置
FONT = 'SimHei'  # 中文字体设置为黑体
COLOR_MAP = {
    'water': (0.255, 0.298, 0.529),
    'land': (135/255, 86/255, 72/255),
    'grid_field': (0.612, 0.702, 0.831),
    'tidal_flat': (246/255, 225/255, 198/255)
}

# 中文到英文的映射
CHINESE_TO_ENGLISH = {
    '水域': 'water',
    '陆地': 'land',
    '养殖区域': 'grid_field',
    '潮滩': 'tidal_flat'
}

def prepare_plot(labels, date, categories_l=['水域', '陆地', '养殖区域', '潮滩']):
    plt.rcParams['font.sans-serif'] = [FONT]
    plt.rcParams['axes.unicode_minus'] = False

    # 确保颜色映射和类别标签只包括你想显示的类别
    actual_colors = [COLOR_MAP[CHINESE_TO_ENGLISH[label]] for label in categories_l if label in CHINESE_TO_ENGLISH]
    actual_labels = [label for label in categories_l if label in CHINESE_TO_ENGLISH]
    cmap = ListedColormap([(1, 1, 1)] + actual_colors)  # 将白色添加为背景色

    # 初始化可视化数组，默认为0（背景色）
    visualization_array = np.zeros_like(next(iter(labels.values())), dtype=int)

    # 根据类别设置颜色索引
    for i, label in enumerate(actual_labels):
        eng_label = CHINESE_TO_ENGLISH[label]
        if eng_label in labels:
            color_index = i + 1  # 从1开始，以0作为背景色
            visualization_array[labels[eng_label]] = color_index

    fig, ax = plt.subplots()
    cax = ax.imshow(visualization_array, cmap=cmap, interpolation='nearest')
    plt.title(date)

    # 创建图例
    patches = [Patch(color=COLOR_MAP[CHINESE_TO_ENGLISH[label]], label=label) for label in actual_labels if CHINESE_TO_ENGLISH[label] in labels]
    plt.legend(handles=patches, bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

    return fig, ax

# 示例数据
labels = {
    'water': np.random.rand(10, 10) > 0.5,
    'land': np.random.rand(10, 10) > 0.5,
    'grid_field': np.random.rand(10, 10) > 0.5,
    'tidal_flat': np.random.rand(10, 10) > 0.5
}  # 随机生成的标签数据
date = "2024-06-01"
fig, ax = prepare_plot(labels, date, categories_l=['水域', '陆地', '养殖区域', '潮滩'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
from matplotlib.patches import Patch

# 超参数设置
FONT = 'SimHei'  # 中文字体设置为黑体
COLOR_MAP = {
    '水域': 1, '陆地': 2, '养殖区域': 3, '潮滩': 4
}

C_MAP_COLORS = [
    (1, 1, 1),  # 白色作为背景
    (0.255, 0.298, 0.529),  # 深蓝
    (135/255, 86/255, 72/255),  # 棕色
    (0.612, 0.702, 0.831),  # 浅蓝
    (246/255, 225/255, 198/255)  # 灰色
]

categories_l = ['水域', '陆地', '养殖区域', '潮滩']  # 显示的类别

def prepare_plot(labels, date):
    plt.rcParams['font.sans-serif'] = [FONT]
    plt.rcParams['axes.unicode_minus'] = False

    cmap = ListedColormap(C_MAP_COLORS)  # 定义颜色映射

    visualization_array = np.zeros_like(next(iter(labels.values())), dtype=int)

    # 根据类别设置颜色索引
    for label in categories_l:
        if label in labels:
            color_index = COLOR_MAP[label]
            visualization_array[labels[label]] = color_index

    fig, ax = plt.subplots()
    cax = ax.imshow(visualization_array, cmap=cmap, interpolation='nearest')
    plt.title(date)

    # 创建图例
    patches = [Patch(color=cmap(COLOR_MAP[label]), label=label) for label in categories_l if label in labels]
    plt.legend(handles=patches, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

    return fig, ax

# 假设labels是真实的标签数据和日期
labels = {'水域': np.random.rand(10, 10) > 0.5, '陆地': np.random.rand(10, 10) > 0.5, '养殖区域': np.random.rand(10, 10) > 0.5, '潮滩': np.random.rand(10, 10) > 0.5}  # 随机生成的标签数据
date = "2024-06-01"
fig, ax = prepare_plot(labels, date)
plt.show()
```

```python
import json
from utils import process_data, get_shape, get_date, prepare_plot
import matplotlib.pyplot as plt
import json
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
from pyproj import Proj, transform
from pyproj import Transformer

# 超参数设置
FONT = 'SimHei'  # 中文字体设置为黑体
COLOR_MAP = {
    'water': 0, 'land': 1, 'grid_field': 2, 'tidal_flat': 3
}
C_MAP_COLORS = [
    (0.255, 0.298, 0.529),  # 深蓝
    (135/255, 86/255, 72/255),  # 棕色
    (0.612, 0.702, 0.831),  # 浅蓝
    (246/255, 225/255, 198/255)  # 灰色
]
LABELS = ['水域', '陆地', '养殖区域', '潮滩']

class Json_loader:
    def __init__(self, filename, transform_filename):
        self.filename = filename
        self.transform_filename = transform_filename
        self.data = self.load_data()
        self.image_shape = get_shape(self.data)
        self.labels = process_data(self.data['shapes'], self.image_shape)
        self.date = get_date(self.filename)
        self.transform = self.load_transform()

    def load_data(self):
        with open(self.filename, 'r') as file:
            return json.load(file)

    def load_transform(self):
        with open(self.transform_filename, 'r') as file:
            return json.load(file)['transform']

    def plot(self):
        fig, ax = self.prepare_plot_with_coords(self.labels, self.date)
        plt.show()

    def plot_save(self, save_path):
        fig, ax = self.prepare_plot_with_coords(self.labels, self.date)
        plt.savefig(save_path)  # 保存图像到指定路径
        plt.close(fig)  # 关闭图形，释放内存

    def prepare_plot_with_coords(self, labels, date):
        plt.rcParams['font.sans-serif'] = [FONT]
        plt.rcParams['axes.unicode_minus'] = False

        # 设置投影转换
        transformer = Transformer.from_crs("epsg:32649", "epsg:4326", always_xy=True)  # 从UTM到WGS84的转换

        cmap = ListedColormap(C_MAP_COLORS)
        visualization_array = np.zeros_like(next(iter(labels.values())), dtype=int)

        for label, mask in labels.items():
            color_index = COLOR_MAP.get(label, 0)  # 未知类别默认为0（水域）
            visualization_array[mask] = color_index

        fig, ax = plt.subplots()
        extent = self.get_extent()
        cax = ax.imshow(visualization_array, cmap=cmap, interpolation='nearest', extent=[0, extent[1] - extent[0], 0, extent[2] - extent[3]])

        cbar = fig.colorbar(cax, ticks=range(len(COLOR_MAP)))
        cbar.set_ticklabels(LABELS)
        plt.title(date)

        # 获取经纬度坐标轴
        left_bottom = transformer.transform(extent[0], extent[3])
        right_top = transformer.transform(extent[1], extent[2])

        ax.set_xlabel('Longitude')
        ax.set_ylabel('Latitude')
        ax.set_xticks(np.linspace(0, extent[1] - extent[0], num=5))  # 设置X轴坐标点
        ax.set_yticks(np.linspace(0, extent[2] - extent[3], num=5))  # 设置Y轴坐标点
        ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda val, pos: f'{left_bottom[0] + val * (right_top[0] - left_bottom[0]) / (extent[1] - extent[0]):.2f}° E'))
        ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda val, pos: f'{left_bottom[1] + val * (right_top[1] - left_bottom[1]) / (extent[3] - extent[2]):.2f}° N'))

        return fig, ax


    def get_extent(self):
        # 转换参数: [c, a, b, f, d, e]
        trans = self.transform
        width, height = self.image_shape[1], self.image_shape[0]
        return [trans[0], trans[0] + trans[1] * width, trans[3] + trans[5] * height, trans[3]]

# 示例使用
filename = r"C:\Users\mikky\Desktop\遥感转可视化项目\全部数据影像\LC08_L1TP_120044_20140203_20200912_02_T1_Water.json"
transform_filename = r"C:\Users\mikky\Desktop\遥感转可视化项目\没啥用的_留个备份\LC08_L1TP_120044_20130827_20200913_02_T1_trans.json"
data = Json_loader(filename, transform_filename)
data.plot()  # 展示图像
# data.plot_save(r"C:\path\to\save\image.png")  # 保存图像到指定路径
```

```python
# import os
# import imageio

# def create_gif_from_jpgs(folder_path, output_filename='output.gif', time=500):  # 将持续时间增加到2秒
#     # 获取指定目录下所有JPG文件的路径
#     files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if f.endswith('.jpg')]
#     # 按文件名排序文件
#     files.sort()
    
#     # 读取图片
#     images = [imageio.imread(file) for file in files]
    
#     # 将图片序列保存为GIF文件
#     imageio.mimsave(output_filename, images, duration=time)

#     return f"GIF created successfully and saved as {output_filename}"

import os
import imageio
from datetime import datetime

def create_gif_from_jpgs(folder_path, output_filename='output.gif', time=500):  # 将持续时间增加到2秒
    # 获取指定目录下所有JPG文件的路径
    files = [f for f in os.listdir(folder_path) if f.endswith('.jpg')]
    
    # 解析文件名中的日期并与文件名一起存储为元组
    file_date_tuples = [(f, datetime.strptime(f.split('_')[3], '%Y%m%d')) for f in files]
    
    # 根据日期从远到近对文件进行排序
    sorted_files = [file for file, _ in sorted(file_date_tuples, key=lambda x: x[1])]
    print(sorted_files,len(sorted_files))
    
    # 读取图片
    images = [imageio.imread(os.path.join(folder_path, file)) for file in sorted_files]
    
    # 将图片序列保存为GIF文件
    imageio.mimsave(output_filename, images, duration=time)

    return f"GIF created successfully and saved as {output_filename}"

```
